```python
import pandas as pd
import numpy as np
```

```python
# df = pd.read_excel(r'coffee_full.xlsx')
df = pd.read_excel (r'coffee_raw2.xlsx')
```

```python
df = df[df['Est. Price'].notna()]
df.head()
```

```python
df['review'].head()
```

# Espresso

```python
df['Espresso'] = df['review'].str.contains('Evaluated as espresso', regex=False)
```

# Extract attributes

```python
df_new = df
```

```python
df['review_list'] = df['review'].str.split('.', n = 10, expand = False)
```

```python
df['review_sentence'] = df['review_list']
```

```python
for i in range(len(df['review_list'])):
    if any(pd.isna(df['review_list'].iloc[0])):
        continue
    elif df['Espresso'].iloc[i] == True:
        df['review_sentence'].iloc[i] = df['review_list'].iloc[i][2]
    else:
        if len(df['review_list'].iloc[i]) == 1:
            df['review_sentence'].iloc[i] = df['review_list'].iloc[i][0]
        else:
            df['review_sentence'].iloc[i] = df['review_list'].iloc[i][1]
```

```python
df['review_attributes_list'] = df['review_sentence'].str.split(',', n = 100, expand = False)
df['review_attributes_split'] = df['review_attributes_list']

df = df.explode('review_attributes_split')
df.reset_index(inplace=True)
```

```python
df['review_attributes_split'] = df['review_attributes_split'].str.strip()
```

```python
counts = df.groupby('review_attributes_split').count().sort_values('coffee', ascending=False)
counts.iloc[:20]
```

# Clean attributes

```python
attributes = df['review_attributes_split'].unique()
attributes
```

```python
# always run with prev cell
attributes = [x.lower() for x in attributes]
print(len(attributes))
attributes = [x.split(' in arom',1)[0] for x in attributes]
attributes = [x.split('-like',1)[0] for x in attributes]
print(len(attributes))
attributes = [x.split(' of ',1)[1] if len(x.split(' of ',1)) == 2 else x.split(' of ',1)[0] for x in attributes]
attributes = [x.split('suggests ',1)[1] if len(x.split('suggests ',1)) == 2 else x.split('suggests ',1)[0] for x in attributes]

attributes_1 = [x.split(' and ',1)[0] for x in attributes]
print(len(attributes_1))
attributes_2 = [x.split(' and ',1)[1] for x in attributes if len(x.split(' and ',1)) == 2]
print(len(attributes_2))

attributes = list(set(attributes_1 + attributes_2))
print(len(attributes))

attributes_1 = [x.split(' or ',1)[0] for x in attributes]
attributes_2 = [x.split(' or ',1)[1] for x in attributes if len(x.split(' or ',1)) == 2]

attributes = list(set(attributes_1 + attributes_2))

attributes_1 = [x.split('; ',1)[0] for x in attributes]
attributes_2 = [x.split('; ',1)[1] for x in attributes if len(x.split('; ',1)) == 2]

attributes = list(set(attributes_1 + attributes_2))

attributes_1 = [x.split(' to ',1)[0] for x in attributes]
attributes_2 = [x.split(' to ',1)[1] for x in attributes if len(x.split(' to ',1)) == 2]

attributes = list(set(attributes_1 + attributes_2))

attributes_1 = [x.split(' with ',1)[0] for x in attributes]
attributes_2 = [x.split(' with ',1)[1] for x in attributes if len(x.split(' with ',1)) == 2]

attributes = list(set(attributes_1 + attributes_2))
print(len(attributes))

attributes = [ x for x in attributes if "acidity" not in x ]
attributes = [ x for x in attributes if "mouthfeel" not in x ]
print(len(attributes))


attributes
```

```python
# list to replace variations
to_replace = {'chocolaty':'chocolate', 
              'passionfruit':'passion fruit',
              'floral':'flower',
              'grapy fruit':'grape',
              'liquorice':'licorice',
              'black-truffle':'black truffle',
              'berries':'berry',
              'lavendar':'lavender',
              'starfruit':'star fruit',
              'port-wine':'port wine',
              'bergamot':'bergamot orange',
              'flowers':'flower',
              'orangy':'orange',
              'pomogranate':'pomegranate',
              'herbal':'herb',
              'blackcurrant':'black currant',
              'hop flower':'hop blossom',
              'pruny':'prune',
              'passionflower':'passion flower',
              'red-currant':'red currant',
              'blackcurrant':'black currant',
              'pasion-fruit':'passion fruit',
              'clover flower':'clover blossom',
              'myrhh':'myrrh',
              'melony':'melon',
              'star-anise':'star anise',
              'cacoa':'cocoa',
              'woody':'wood',
              'berries':'berry',
              'fruits':'fruit',
              'fruity':'fruit'
             }
```

```python
# replace variations in reviews
for review in df['review']:
    for orig, replacement in to_replace.items():
        review = review.replace(orig,replacement)
```
