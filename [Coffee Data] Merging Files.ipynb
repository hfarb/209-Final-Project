```python
import pandas as pd
import numpy as np
```

```python
prev = pd.read_excel(r'coffee_full.xlsx')
aromas_only = pd.read_excel(r'coffee_aromas_only.xlsx')
```

```python
aromas_only = aromas_only.iloc[:,1:]
```

```python
aromas_only.reset_index(inplace=True)
aromas_only.head()
```

```python
coffee_full2 = prev.merge(aromas_only, on=('index'), suffixes=('_prev', '_aromas'))
```

```python
# coffee_full2.to_excel('coffee_full2.xlsx',index = True)
```

```python
coffee_full2.head(1)
```

```python
df_temp = coffee_full2.iloc[:,np.r_[0,18,33:245]]
df_temp.head(5)
```

```python
df_dropped = df_temp.drop_duplicates(subset ="index", keep = False, inplace = False) 
```

```python
df_dropped
```

```python
aroma_counts = df_dropped.groupby('Coffee Origin_last').sum()
aroma_counts = aroma_counts.iloc[:,1:]
aroma_counts = aroma_counts.iloc[:,np.r_[0,2:212]]
aroma_counts
```

```python
aroma_counts.to_excel('coffee_aromas_counts.xlsx',index = True)
```

```python
aroma_percentages = aroma_counts.div(aroma_counts.sum(axis=1), axis=0)
```

```python
# aroma_counts['total'] = aroma_counts.sum(axis=1)
```

```python
aroma_percentages.head()
```

```python
aroma_percentages.to_excel('coffee_aromas_percentages.xlsx',index = True)
```

```python
#sort attributes further (layer_1, layer_2, etc.)
#pick top 5 counts

df_5 = pd.DataFrame(columns = ['country', 'attribute', 'count'])

for i in range(len(aroma_counts)):
    df_sub = pd.DataFrame() 
    df_sub['country'] = [aroma_counts.iloc[i,].name]*5
    df_sub['attribute'] = aroma_counts.iloc[i,].sort_values(ascending = False)[0:5].index.to_list()
    df_sub['count'] = aroma_counts.iloc[i,].sort_values(ascending = False)[0:5].to_list()
    top5_total = sum(aroma_counts.iloc[i,].sort_values(ascending = False)[0:5].to_list())
    df_sub['percentage'] = [x/top5_total for x in df_sub['count'].to_list()]
    df_5 = df_5.append(df_sub)
```

```python
df_5
```

```python
df_5.to_excel('coffee_aromas_top5.xlsx',index = True)
```
